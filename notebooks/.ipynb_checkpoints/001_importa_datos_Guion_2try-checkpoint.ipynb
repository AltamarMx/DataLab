```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
```

```python
f = 'https://www.ier.unam.mx/~gbv/DataLab/H2O.csv'
agua = pd.read_csv(f,index_col=0,parse_dates=True,names=['t','consumo'],skiprows=1)
agua.plot(figsize=(12,4))
```

```python
agua.resample('D').sum().plot(figsize=(12,4),kind='bar')
```

## Prompt

```python
'''
Tengo un dataframe llamado agua con un ndice que es datetime con datos cada hora en una columna llamada consumo.


Quiero calcular el promedio de consumo  de la serie temporal y luego detectar donde el consumo diario sea mayor al promedio diario de la serie temporal.
'''
```

```python

# Asegrate de que la columna de ndice sea de tipo datetime
# agua.index = pd.to_datetime(agua.index)

# Calcular el promedio de consumo de la serie temporal
promedio_consumo = agua['consumo'].mean()

# Agrupar los datos por da y calcular el promedio diario de consumo
promedio_diario = agua['consumo'].resample('D').mean()

# Detectar dnde el consumo diario es mayor que el promedio diario
consumo_mayor_que_promedio = promedio_diario[promedio_diario > promedio_consumo]

# Imprimir el promedio de consumo y las fechas donde el consumo diario es mayor que el promedio
print("Promedio de consumo:", promedio_consumo)
print("Fechas con consumo diario mayor que el promedio:")
print(consumo_mayor_que_promedio)
```

## Prompt

```python
'''
que visualizacin me propones de la serie temporal para localizar los das con consumo diario mayor que el promedio ? 


'''
```

```python
# import matplotlib.pyplot as plt

# Crear un grfico de lneas de consumo diario
plt.figure(figsize=(12, 6))
plt.plot(promedio_diario.index, promedio_diario.values, label='Consumo Diario')
plt.axhline(y=promedio_consumo, color='red', linestyle='--', label='Promedio Diario')

# Resaltar los das con consumo mayor que el promedio
plt.fill_between(promedio_diario.index, promedio_diario.values, where=promedio_diario > promedio_consumo, color='lightcoral', alpha=0.5, label='Das con Consumo Mayor')

# Configuracin del grfico
plt.xlabel('Fecha')
plt.ylabel('Consumo')
plt.title('Consumo Diario de Agua')
plt.legend()
plt.grid(True)

# Mostrar el grfico
plt.show()
```

```python
# import pandas as pd
# import matplotlib.pyplot as plt

# Asegrate de que el ndice del DataFrame sea de tipo datetime
agua.index = pd.to_datetime(agua.index)

# Calcular el promedio de consumo por da de la semana
promedio_por_dia_semana = agua['consumo'].groupby(agua.index.dayofweek).mean()

# Calcular el promedio diario de consumo
promedio_diario = agua['consumo'].resample('D').mean()

# Crear una serie para almacenar las fechas en que el consumo es mayor que el promedio por da de la semana
fechas_destacadas = []

# Iterar a travs de los datos y detectar las fechas destacadas
for fecha, consumo in zip(promedio_diario.index, promedio_diario.values):
    dia_semana = fecha.dayofweek
    promedio_dia_semana = promedio_por_dia_semana[dia_semana]
    if consumo > promedio_dia_semana:
        fechas_destacadas.append(fecha)

# Crear un grfico de barras de consumo diario
plt.figure(figsize=(12, 6))
plt.bar(promedio_diario.index, promedio_diario.values, label='Consumo Diario', color='lightblue')

# Resaltar las fechas en que el consumo diario es mayor que el promedio por da de la semana
plt.bar([fecha.date() for fecha in fechas_destacadas], [promedio_diario.loc[fecha] for fecha in fechas_destacadas], color='salmon', label='Das con Consumo Mayor')

# Configuracin del grfico
plt.xlabel('Fecha')
plt.ylabel('Consumo')
plt.title('Consumo Diario de Agua')
plt.legend()
plt.grid(True)

# Mostrar el grfico
plt.show()
```

Tomo la opci'on de boxplot

```python
consumo_diario = agua['consumo'].resample('D').sum()

# import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.boxplot(consumo_diario, vert=False)
plt.title('Diagrama de Caja del Consumo Diario de Agua')
plt.xlabel('Consumo Diario')
plt.show()
```

## Prompt

```python
'''
alrededor de 8 puntos quedaron fuera de los bigotes, como identifico esos das de mi serie temporal ? 

'''
```

```python

# Crear el diagrama de caja y obtener los valores atpicos
boxplot = plt.boxplot(consumo_diario, vert=False)
valores_atipicos = [flier.get_xdata() for flier in boxplot['fliers']]

# Identificar los das correspondientes a los valores atpicos
dias_atipicos = []
for valor_atipico in valores_atipicos:
    for i, consumo_dia in enumerate(consumo_diario):
        if consumo_dia in valor_atipico:
            dia_atipico = consumo_diario.index[i].strftime('%Y-%m-%d')
            dias_atipicos.append(dia_atipico)
print("Das con valores atpicos:")
for dia in dias_atipicos:
    print(dia)
```

## Prompt

```python
'''
ayudame a graficar esos das en una grafica de consumo diario usando barras e incluye el umbral con una lnea roja el valor donde el consumo diario comienza a considerarse atpico de acuerdo a la boxplot.


'''
```

```python
# import matplotlib.pyplot as plt
# import numpy as np

# Crear la grfica de barras del consumo diario
plt.figure(figsize=(12, 6))
plt.bar(consumo_diario.index, consumo_diario, label='Consumo Diario', color='lightblue')

# Identificar el umbral de valores atpicos segn el boxplot
umbral_atipico = np.percentile(consumo_diario, 75) + 1.5 * (np.percentile(consumo_diario, 75) - np.percentile(consumo_diario, 25))

# Agregar una lnea roja para el umbral de valores atpicos
plt.axhline(y=umbral_atipico, color='red', linestyle='--', label='Umbral de Valores Atpicos')

# Identificar y marcar los das con valores atpicos
dias_atipicos_indices = [consumo_diario.index.get_loc(dia) for dia in dias_atipicos]
plt.bar([consumo_diario.index[i] for i in dias_atipicos_indices], [consumo_diario[i] for i in dias_atipicos_indices], color='salmon', label='Das Atpicos')

# Configuracin del grfico
plt.xlabel('Fecha')
plt.ylabel('Consumo')
plt.title('Consumo Diario de Agua con Valores Atpicos')
plt.legend()
plt.grid(True)

# Mostrar el grfico
plt.show()
```

```python
'''
Ahora explicame como sera detectar anomalias usando histogramas
'''
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(consumo_diario, bins=20, color='lightblue', edgecolor='black')
plt.title('Histograma del Consumo Diario de Agua')
plt.xlabel('Consumo Diario')
plt.ylabel('Frecuencia')
plt.show()
```

```python
percentil_inferior = np.percentile(consumo_diario, 5)
percentil_superior = np.percentile(consumo_diario, 95)

# Identificar das con anomalas
dias_anomalos = consumo_diario[(consumo_diario < percentil_inferior) | (consumo_diario > percentil_superior)].index
```
